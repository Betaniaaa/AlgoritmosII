Ejercicio 1: Clase CocheDeCarreras


```python

class CocheDeCarreras: #definicion de la clase
    def __init__(self, nombre, velocidad_maxima): #atributos
        self.nombre = nombre
        self.velocidad_maxima = velocidad_maxima
        self.velocidad_actual = 0
        self.combustible = 100


#Metodos
#Acelerar (incremento)

    def acelerar(self, incremento):
        if self.combustible == 0:
            print(f"Error: {self.nombre} no tiene combustible para acelerar.")
            return

        if self.velocidad_actual + incremento > self.velocidad_maxima:
            self.velocidad_actual = self.velocidad_maxima
        else:
            self.velocidad_actual += incremento

        # Reducir combustible en función de la velocidad incrementada
        self.combustible -= incremento // 10
        if self.combustible < 0:
            self.combustible = 0

        # Verificación de velocidad peligrosa
        if self.velocidad_actual > 0.9 * self.velocidad_maxima:
            print("¡Cuidado! Velocidad peligrosa")

#frenar (decremento)
    def frenar(self, decremento):
        if self.velocidad_actual - decremento < 0:
            self.velocidad_actual = 0
        else:
            self.velocidad_actual -= decremento
#repostar (cantidad)
    def repostar(self, cantidad):
        self.combustible += cantidad
        if self.combustible > 100:
            self.combustible = 100
#estado actual
    def estado(self):
        print(f"\n--- Estado del coche ---")
        print(f"Nombre: {self.nombre}")
        print(f"Velocidad Actual: {self.velocidad_actual} km/h")
        print(f"Combustible: {self.combustible}%")
        print("-------------------------\n")

#  interacción con el usuario
def interactuar_con_usuario():
    print("Este es un simulador de coche de carreras!")
    nombre = input("Introduce el nombre de tu coche: ")
    velocidad_maxima = int(input(f"Introduce la velocidad máxima de {nombre} en km/h: "))
    coche = CocheDeCarreras(nombre, velocidad_maxima)

    while True:
        print("\n¿Qué te gustaría hacer ahora?")
        print("1. Acelerar")
        print("2. Frenar")
        print("3. Repostar combustible")
        print("4. Ver estado del coche")
        print("5. Salir del simulador")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == "1":
            if coche.combustible == 0:
                print("No tienes suficiente combustible para acelerar.")
            else:
                incremento = int(input(f"Introduce el incremento de velocidad (máximo {coche.velocidad_maxima} km/h): "))
                coche.acelerar(incremento)
        elif opcion == "2":
            decremento = int(input("Introduce el valor para frenar (en km/h): "))
            coche.frenar(decremento)
        elif opcion == "3":
            cantidad = int(input("¿Cuánto combustible quieres repostar? (máximo 100%): "))
            coche.repostar(cantidad)
        elif opcion == "4":
            coche.estado()
        elif opcion == "5":
            print("¡Gracias por jugar! ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, selecciona una opción entre 1 y 5.")

# Ejecutar la función de interacción con el usuario
interactuar_con_usuario()

```

Ejercicio 2: Clase Robot

```python
class Robot:
    def __init__(self, nombre):
        self.nombre = nombre
        self.bateria = 100
        self.posicion = (0, 0)

    def mover(self, direccion, distancia):
        if self.bateria <= 0:
            print("Batería agotada. No se puede mover.")
            return

        movimientos = {"arriba": (0, 1), "abajo": (0, -1), "izquierda": (-1, 0), "derecha": (1, 0)}
        if direccion not in movimientos:
            raise DireccionInvalidaError("Dirección inválida.")

        x, y = self.posicion
        dx, dy = movimientos[direccion]
        nueva_pos = (x + dx * distancia, y + dy * distancia)

        if nueva_pos[0] < 0 or nueva_pos[1] < 0:
            print("Movimiento inválido. No se permiten posiciones negativas.")
            return

        self.posicion = nueva_pos
        self.bateria -= distancia
        print(f"{self.nombre} se movió a {self.posicion}. Batería: {self.bateria}%")

    def cargar_bateria(self, cantidad):
        self.bateria = min(100, self.bateria + cantidad)
        print(f"Batería cargada: {self.bateria}%")

    def disparar(self):
        if self.bateria < 5:
            print("Batería insuficiente para disparar.")
            return
        self.bateria -= 5
        print(f"{self.nombre} disparó un láser. Batería: {self.bateria}%")

    def estado(self):
        print(f"{self.nombre} - Posición: {self.posicion}, Batería: {self.bateria}%")

# Ejemplo de uso
robot = Robot("Bett")
robot.estado()
robot.mover("arriba", 10)
robot.mover("izquierda", 5)
robot.disparar()
robot.cargar_bateria(20)
robot.estado()
```

Ejercicio 3: Clase JuegoDeCartas


```python
import random

# Excepción
class NoHaySuficientesCartasError(Exception):
    pass

class JuegoDeCartas:
    def __init__(self):
        self.mazo = []  # Mazo vacío
        self.jugadores = {}  # Diccionario para almacenar las manos de los jugadores

    def agregar_carta(self, valor, pinta):
        """Añade una carta al mazo"""
        carta = (valor, pinta)
        self.mazo.append(carta)

    def repartir(self, jugador, cantidad):
        """Reparte cartas a un jugador. Lanza error si no hay suficientes cartas."""
        if len(self.mazo) < cantidad:
            raise NoHaySuficientesCartasError(f"No hay suficientes cartas en el mazo para repartir {cantidad} cartas.")

        # Repartir las cartas al jugador
        mano = []
        for _ in range(cantidad):
            carta = self.mazo.pop()  # Sacamos la última carta del mazo
            mano.append(carta)

        # Si el jugador no existe, creamos su mano
        if jugador not in self.jugadores:
            self.jugadores[jugador] = []

        self.jugadores[jugador].extend(mano)

    def mostrar_mano(self, jugador):
        """Muestra las cartas que tiene un jugador"""
        if jugador not in self.jugadores:
            print(f"El jugador {jugador} no tiene cartas.")
            return
        print(f"Mano de {jugador}: {self.jugadores[jugador]}")

    def mezclar_mazo(self):
        """Mezcla el mazo de cartas"""
        random.shuffle(self.mazo)

# Función para interactuar con el usuario
def interactuar_con_usuario():
    juego = JuegoDeCartas()

    while True:
        print("\n¿Qué te gustaría hacer ahora?")
        print("1. Agregar carta al mazo")
        print("2. Repartir cartas a un jugador")
        print("3. Mostrar mano de un jugador")
        print("4. Mezclar el mazo")
        print("5. Salir del juego")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == "1":
            valor = input("Introduce el valor de la carta (Ej. As, 2, 3, ..., Rey): ")
            pinta = input("Introduce la pinta de la carta (Ej. Corazones, Diamantes, Tréboles, Picas): ")
            juego.agregar_carta(valor, pinta)
            print(f"Carta {valor} de {pinta} agregada al mazo.")

        elif opcion == "2":
            jugador = input("Introduce el nombre del jugador: ")
            cantidad = int(input("¿Cuántas cartas quieres repartir? "))
            try:
                juego.repartir(jugador, cantidad)
                print(f"{cantidad} cartas repartidas a {jugador}.")
            except NoHaySuficientesCartasError as e:
                print(e)

        elif opcion == "3":
            jugador = input("Introduce el nombre del jugador: ")
            juego.mostrar_mano(jugador)

        elif opcion == "4":
            juego.mezclar_mazo()
            print("El mazo ha sido mezclado.")

        elif opcion == "5":
            print("¡Gracias por jugar! ¡Hasta luego!")
            break

        else:
            print("Opción no válida. Por favor, selecciona una opción entre 1 y 5.")

# Ejecutar la función de interacción con el usuario
interactuar_con_usuario()
```
